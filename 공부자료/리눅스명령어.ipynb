## 반드시 왼쪽 작업디렉토리를 모니터링하면서 진행

```python
# !와 % 차이점

느낌표(!)는 조회, 생성 같이 한번만 실행할때 쓰는 magic line 언어이다. 단, 경로 이동은 적용되지 않는다.
퍼센테이지(%)는 실제 경로 이동을 반영할때 주로 쓰이며, 모든 쉘 언어 사용 가능
```

```python
%cd ..  # 상위 디렉토리
```

```python
# 현재 작업 경로 출력

!pwd
```

```python
# 디렉토리 만들기
!mkdir test
```

```python
# 현재 작업 경로에 있는 모든 파일 확인
!ls
```

```python
# 폴더 이동
%cd test
```

```python
# 파일 한번 만들어보기(이론 설명 후 실습)
alpaco폴더 만들어보기 (좀있다가)

!mkdir alpaco
```

```python
!pwd
```

```python
# 모든 파일의 모든 정보 보기
!ls -al    # drwxr-xr-x : 파일 권한  /  r : read  /  w : write
```

```python
!chmod 007 osh.txt
```

```python
# 디렉토리 만들기
!mkdir a
!mkdir b
!mkdir c
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# a 디렉토리의 권한 수정하기
!chmod 700 a
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# a 파일을 상위 디렉토리로 옮기기
!mv ./a ../        # mv : move -> a라는 파일을 부모 폴더로 옮기겠다
```

```python
# c 파일을 상위 디렉토리로 복사하기
!cp -r ./c ../    # cp : copy
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# c 파일을 b 디렉토리로 복사하기
!cp -r ./c ./b
```

```python
# b 디렉토리 안에 있는 정보 보기
!ls -al ./b
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# 파일 삭제하기, 오류가 날 것 이다.
!rm b    # rm : remove
```

```python
# 폴더 또는 그 폴더 안에 있는 파일도 삭제할 때 -r 옵션을 사
!rm -r b
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# 디스크 용량 확인
!df -h
```

```python
# cpu, ram 정보 확인 (계속 상태 유지되니까 앞에 정지버튼 누르기)
!top    # PID : 프로그램 순번
```
